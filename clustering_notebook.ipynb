```python
import numpy as np
import pandas as pd
import scanpy as sp
```

```python
adata = sp.read_csv('human_pancreas_cells.csv.gz', delimiter=',', first_column_names=None, dtype='float32')
```

```python
adata=adata.transpose()
```

```python
sp.pl.highest_expr_genes(adata, n_top=20, )
```

```python
sp.pp.filter_cells(adata, min_genes=200)
sp.pp.filter_genes(adata, min_cells=3)
```

```python
sp.pp.normalize_total(adata, target_sum=1e4)

sp.pp.log1p(adata)
sp.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
```

```python
sp.tl.pca(adata, svd_solver='arpack')
sp.pl.pca(adata, color='CST3')
```

```python
sp.pp.neighbors(adata, n_neighbors=10, n_pcs=40)

sp.tl.umap(adata)
```

```python
sp.tl.leiden(adata)
```

```python
sp.pl.umap(adata, color=['leiden])
```
